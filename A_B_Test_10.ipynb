```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest
import itertools

# --- Annahme: eniac_a, eniac_b, eniac_c, eniac_d sind schon geladen ---

# 1. Klicks für sichtbaren Button extrahieren
def extract_button_clicks(df):
    return df[(df["Name"].isin(["SHOP NOW", "SEE DEALS"])) & (df["Visible?"] == True)]["No. clicks"].sum()

# 2. Versionen in Dictionary
versions = {
    "A (White SHOP NOW)": eniac_a,
    "B (Red SHOP NOW)": eniac_b,
    "C (White SEE DEALS)": eniac_c,
    "D (Red SEE DEALS)": eniac_d
}

# 3. Gesamtseitenaufrufe & pro Version (Annahme: gleichverteilt)
total_views = 50061
views_per_version = total_views / 4

# 4. CTR berechnen
ctr_results = {}
clicks_results = {}
for name, df in versions.items():
    clicks = extract_button_clicks(df)
    ctr = clicks / views_per_version
    clicks_results[name] = clicks
    ctr_results[name] = ctr

print("CTR Results:")
for k, v in ctr_results.items():
    print(f"{k}: CTR = {v:.4%} ({clicks_results[k]} clicks / {int(views_per_version)} views)")

# 5. Gewinner finden
winner = max(ctr_results, key=ctr_results.get)
print(f"\nWinner: {winner} with CTR {ctr_results[winner]:.4%}")

# 6. Paarweise Z-Tests für Proportionen
print("\nPairwise significance tests (Z-test for proportions):")

alpha = 0.05
pairs = list(itertools.combinations(versions.keys(), 2))
for v1, v2 in pairs:
    clicks = [clicks_results[v1], clicks_results[v2]]
    views = [views_per_version, views_per_version]
    stat, p_val = proportions_ztest(count=clicks, nobs=views, alternative='two-sided')
    signif = "SIGNIFICANT" if p_val < alpha else "not significant"
    print(f"{v1} vs {v2}: Z = {stat:.3f}, p = {p_val:.4f} -> {signif}")

# 7. Balkendiagramm plotten
plt.figure(figsize=(9, 5))
bars = plt.bar(ctr_results.keys(), [v * 100 for v in ctr_results.values()],
               color=['gray', 'red', 'lightblue', 'darkred'])
plt.title('Click-Through Rate (CTR) by Homepage Version')
plt.ylabel('CTR (%)')
plt.xlabel('Homepage Version')
plt.ylim(0, max(ctr_results.values()) * 100 + 1)
plt.grid(axis='y', linestyle='--', alpha=0.5)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height + 0.1,
             f"{height:.2f}%",
             ha='center',
             va='bottom',
             fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Annahme: eniac_a, eniac_b, eniac_c, eniac_d sind DataFrames, bereits geladen

total_visits = 50061
visits_per_version = total_visits / 4  # Gleichverteilung

# Funktion: Klicks auf sichtbare Buttons extrahieren
def extract_button_clicks(df):
    return df[(df["Name"].isin(["SHOP NOW", "SEE DEALS"])) & (df["Visible?"] == True)]["No. clicks"].sum()

# Klicks pro Version berechnen
clicks_a = extract_button_clicks(eniac_a)
clicks_b = extract_button_clicks(eniac_b)
clicks_c = extract_button_clicks(eniac_c)
clicks_d = extract_button_clicks(eniac_d)

# Nicht-Klicks berechnen
no_clicks_a = visits_per_version - clicks_a
no_clicks_b = visits_per_version - clicks_b
no_clicks_c = visits_per_version - clicks_c
no_clicks_d = visits_per_version - clicks_d

# Kontingenztabelle für Chi-Quadrat-Test erstellen
contingency_table = [
    [clicks_a, no_clicks_a],
    [clicks_b, no_clicks_b],
    [clicks_c, no_clicks_c],
    [clicks_d, no_clicks_d]
]

print("Contingency Table (Clicks / No Clicks):")
print(pd.DataFrame(contingency_table, columns=["Clicks", "No Clicks"], index=["A", "B", "C", "D"]))

# Chi-Quadrat-Test ausführen
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

# Interpretation
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis. There is a statistically significant difference in CTR between versions.")
else:
    print("Result: Fail to reject the null hypothesis. No statistically significant difference detected.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Werte aus vorherigem Code:
total_visits = 50061
visits_per_version = int(total_visits / 4)

clicks = {
    "Version A": int(clicks_a),
    "Version B": int(clicks_b),
    "Version C": int(clicks_c),
    "Version D": int(clicks_d)
}

# CTRs als Wahrscheinlichkeiten
probs = {k: v/visits_per_version for k, v in clicks.items()}

x = np.arange(0, visits_per_version + 1)  # mögliche Klickanzahlen

plt.figure(figsize=(10,6))

for version in clicks.keys():
    p = probs[version]
    # Binomialverteilung für jeden Wert x berechnen
    pmf = binom.pmf(x, visits_per_version, p)
    # Normieren für bessere Sichtbarkeit (max auf 1 skalieren)
    pmf_norm = pmf / pmf.max()
    plt.plot(x / visits_per_version * 100, pmf_norm, label=f"{version} (p={p:.4f})")

plt.xlabel("Click-Through Rate (%)")
plt.ylabel("Normalized Probability")
plt.title("Binomial Distribution of Click-Through Rates per Version")
plt.legend()
plt.grid(True)
plt.show()
```

```python
Goal:

Since we only have aggregated data (total clicks and visits per version), we approximate the distribution of clicks per version using the Binomial distribution.

Plot the probability mass function (PMF) for each version's clicks, normalized for easier comparison.

The x-axis shows CTR in percent, the y-axis shows the relative probability of observing that CTR.
```

Explanation

We model the clicks for each version as a Binomial distribution with
𝑛
=
n= number of visits per version and
𝑝
=
p= click-through rate (CTR).

For each possible number of clicks
𝑘
k, we calculate the probability of observing exactly
𝑘
k clicks (the PMF).

We normalize each PMF curve to a maximum height of 1 to make them comparable visually.

The x-axis shows the CTR in percent, and the y-axis shows the relative probability of that CTR.

This helps visualize how spread out or tight the expected CTRs are for each version and whether the versions are clearly different in their click behavior.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Values from previous code:
total_visits = 50061
visits_per_version = int(total_visits / 4)

clicks = {
    "Version A": int(clicks_a),
    "Version B": int(clicks_b),
    "Version C": int(clicks_c),
    "Version D": int(clicks_d)
}

# CTR probabilities
probs = {k: v / visits_per_version for k, v in clicks.items()}

x = np.arange(0, visits_per_version + 1)  # possible number of clicks

plt.figure(figsize=(10,6))

for version in clicks.keys():
    p = probs[version]
    # Calculate Binomial PMF for each x
    pmf = binom.pmf(x, visits_per_version, p)
    # Normalize PMF to max = 1 for better visibility
    pmf_norm = pmf / pmf.max()
    plt.plot(x / visits_per_version * 100, pmf_norm, label=f"{version} (p={p:.4f})")

plt.xlabel("Click-Through Rate (%)")
plt.ylabel("Normalized Probability")
plt.title("Binomial Distribution of Click-Through Rates per Version")
plt.legend()
plt.grid(True)
plt.show()
```

3. Erklärung auf Deutsch

Für jedes Paar von Versionen wird geprüft, ob die Differenz der Klickrate signifikant ist.

Der Z-Test basiert auf der Normalverteilung, da wir mit großen Stichproben arbeiten (Central Limit Theorem).

Der p-Wert zeigt die Wahrscheinlichkeit, dass ein Unterschied dieser Größe zufällig entstehen kann.

Wenn p < 0,05 (Alpha), lehnen wir die Nullhypothese ab und sagen, dass der Unterschied signifikant ist.

```python
Wir führen paarweise Z-Tests durch, um zu prüfen, ob die Unterschiede der CTRs zwischen den Versionen signifikant sind.
```

```python
import numpy as np
from scipy.stats import norm

# Klicks und Visits pro Version (aus vorheriger Berechnung)
visits_per_version = int(total_visits / 4)
clicks_dict = {
    "A": clicks_a,
    "B": clicks_b,
    "C": clicks_c,
    "D": clicks_d
}

def z_test_proportions(x1, n1, x2, n2):
    p1 = x1 / n1
    p2 = x2 / n2
    p_combined = (x1 + x2) / (n1 + n2)
    numerator = p1 - p2
    denominator = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
    z = numerator / denominator
    p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed test
    return z, p_value

# Paare zum Testen
pairs = [("A","B"), ("A","C"), ("A","D"), ("B","C"), ("B","D"), ("C","D")]

alpha = 0.05

print("Pairwise Z-tests for CTR differences:\n")
for v1, v2 in pairs:
    z, p_val = z_test_proportions(clicks_dict[v1], visits_per_version, clicks_dict[v2], visits_per_version)
    result = "Reject H0" if p_val < alpha else "Fail to reject H0"
    print(f"{v1} vs {v2}: Z = {z:.3f}, p-value = {p_val:.4f} -> {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Klicks und Visits (Beispieldaten, bitte mit deinen Werten ersetzen)
total_visits = 50061
visits_per_version = int(total_visits / 4)

clicks_dict = {
    "A": clicks_a,
    "B": clicks_b,
    "C": clicks_c,
    "D": clicks_d
}

def z_test_proportions(x1, n1, x2, n2):
    p1 = x1 / n1
    p2 = x2 / n2
    p_combined = (x1 + x2) / (n1 + n2)
    numerator = p1 - p2
    denominator = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
    z = numerator / denominator
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

versions = ["A", "B", "C", "D"]
n = len(versions)

# Matrices für Z-Werte und p-Werte
z_matrix = np.zeros((n, n))
p_matrix = np.ones((n, n))

# Paarweise Tests füllen
for i in range(n):
    for j in range(i+1, n):
        z, p = z_test_proportions(clicks_dict[versions[i]], visits_per_version,
                                  clicks_dict[versions[j]], visits_per_version)
        z_matrix[i, j] = z
        z_matrix[j, i] = -z  # Z-Wert spiegelt sich um die Diagonale
        p_matrix[i, j] = p
        p_matrix[j, i] = p

# Plot Heatmap der p-Werte (signifikante Werte farbig)
fig, ax = plt.subplots(figsize=(8,6))

im = ax.imshow(p_matrix, cmap="coolwarm_r", vmin=0, vmax=0.1)

# Achsenbeschriftung
ax.set_xticks(np.arange(n))
ax.set_yticks(np.arange(n))
ax.set_xticklabels(versions)
ax.set_yticklabels(versions)

# Werte in Zellen anzeigen (p-Werte und Signifikanz markieren)
for i in range(n):
    for j in range(n):
        if i == j:
            text = "-"
        else:
            text = f"{p_matrix[i,j]:.3f}"
            if p_matrix[i,j] < 0.05:
                text += "*"
        ax.text(j, i, text, ha="center", va="center", color="black")

plt.colorbar(im, label="p-value")
plt.title("Pairwise Z-Test p-values ( * = significant at alpha=0.05 )")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Annahme: du hast die DataFrames eniac_a, eniac_b, eniac_c, eniac_d bereits geladen ---

# Beispiel: Wir filtern aus jeder Tabelle den Button-Row heraus
def get_clicks(df, button_names):
    # button_names ist Liste der möglichen Namen, z.B. ['SHOP NOW', 'SEE DEALS']
    # Wir suchen Zeilen, die einen der Button-Namen enthalten
    mask = df['Name'].isin(button_names)
    clicks = df.loc[mask, 'No. clicks'].sum()  # summe falls mehrere vorkommen
    return clicks

# Button-Namen (aus deinem Setup, anpassen falls nötig)
button_names_a = ['SHOP NOW']
button_names_b = ['SHOP NOW']
button_names_c = ['SEE DEALS']
button_names_d = ['SEE DEALS']

clicks_a = get_clicks(eniac_a, button_names_a)
clicks_b = get_clicks(eniac_b, button_names_b)
clicks_c = get_clicks(eniac_c, button_names_c)
clicks_d = get_clicks(eniac_d, button_names_d)

# Gesamtbesuche (gegeben, oder man kann sie auch aus Logs nehmen)
total_visits = 50061
visits_per_version = int(total_visits / 4)  # gleichmäßig verteilt

clicks_dict = {
    "A": clicks_a,
    "B": clicks_b,
    "C": clicks_c,
    "D": clicks_d
}

# CTR und Konfidenzintervall Funktion
def proportion_confint(x, n, alpha=0.05):
    p = x / n
    z = norm.ppf(1 - alpha/2)
    se = np.sqrt(p * (1 - p) / n)
    lower = p - z * se
    upper = p + z * se
    return p, lower, upper

# CTRs und CI berechnen
versions = ["A", "B", "C", "D"]
ctrs = []
ci_lowers = []
ci_uppers = []

for v in versions:
    p, lower, upper = proportion_confint(clicks_dict[v], visits_per_version)
    ctrs.append(p * 100)       # in Prozent
    ci_lowers.append((p - lower) * 100)
    ci_uppers.append((upper - p) * 100)

# Paarweise Z-Test Funktion
def z_test_proportions(x1, n1, x2, n2):
    p1 = x1 / n1
    p2 = x2 / n2
    p_combined = (x1 + x2) / (n1 + n2)
    numerator = p1 - p2
    denominator = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
    z = numerator / denominator
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

# P-Value Matrix füllen
n = len(versions)
p_matrix = np.ones((n, n))

for i in range(n):
    for j in range(i+1, n):
        _, p = z_test_proportions(clicks_dict[versions[i]], visits_per_version,
                                  clicks_dict[versions[j]], visits_per_version)
        p_matrix[i, j] = p
        p_matrix[j, i] = p

alpha = 0.05

# Heatmap plotten mit Signifikanz-Hervorhebung
fig, ax = plt.subplots(figsize=(8,6))
cmap = plt.cm.coolwarm_r
im = ax.imshow(p_matrix, cmap=cmap, vmin=0, vmax=0.1)

ax.set_xticks(np.arange(n))
ax.set_yticks(np.arange(n))
ax.set_xticklabels(versions)
ax.set_yticklabels(versions)

for i in range(n):
    for j in range(n):
        if i == j:
            text = "-"
            ax.text(j, i, text, ha="center", va="center", color="black")
        else:
            p_val = p_matrix[i, j]
            text = f"{p_val:.3f}"
            if p_val < alpha:
                text += "*"
                rect = plt.Rectangle((j-0.5, i-0.5), 1, 1, linewidth=2, edgecolor='green', facecolor='none')
                ax.add_patch(rect)
            ax.text(j, i, text, ha="center", va="center", color="black")

cbar = plt.colorbar(im, ax=ax)
cbar.set_label('p-value')
# Linie bei alpha
cbar.ax.hlines(y=alpha / 0.1 * cbar.vmax, xmin=0, xmax=1, colors='green', linestyles='dashed', linewidth=2)
cbar.ax.text(0.5, alpha / 0.1 * cbar.vmax + 0.01, f'α = {alpha}', color='green', ha='center', va='bottom')

plt.title("Pairwise Z-Test p-values with significance highlight (green border and *)")
plt.show()

# CTR Balkendiagramm mit Fehlerbalken
plt.figure(figsize=(8,6))
plt.bar(versions, ctrs, yerr=[ci_lowers, ci_uppers], capsize=8, color=['gray', 'red', 'gray', 'red'])
plt.ylabel('Click-Through Rate (%)')
plt.title('Click-Through Rates with 95% Confidence Intervals')
plt.grid(axis='y')
plt.show()
```
